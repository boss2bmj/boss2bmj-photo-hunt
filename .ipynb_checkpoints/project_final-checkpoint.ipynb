In this step, we make the photo to be blurred because we want to eiliminate noise error to make it more accurate in term of comparing points between two photo.

### In this step, we divide the photo the we blured into two picture(left side and right side).

### How we find the different point between two pictures.

In this step, we use vector to find the different point between 2 pictures.

```python
#we use vector find the different point between 2 picturs that we seperates above
def find_different_point(first_pic,second_pic):
    new_image = np.empty(first_pic)
    keep = []
    data = np.zeros((1,2)) #to make an array
    keep_value_image = np.empty(first_pic)
    for i in range(first_blurred_half.shape[0]): #first dimension of pic
        for j in range(first_blurred_half.shape[1]): #second dimension of pic
            rl = first_blurred_half[i,j,0] #red,left picture
            gl = first_blurred_half[i,j,1] #greed
            bl = first_blurred_half[i,j,2] #blue
            rr = second_blurred_half[i,j,0]
            gr = second_blurred_half[i,j,1]
            br = second_blurred_half[i,j,2]
            distance = ((rl-rr)**2 + (bl-br)**2 + (gl-gr)**2)**0.5 #to find distance to compare between 2 vectors
            if distance > 0.45: # here, if the distance different too much, so it is not the same
                #In this case, we have to find the best value to determine whether the color on that point are the same or not,
                #in this pic we choose 0.45(it depends on picture)
                new = np.array((j,i))
                data = np.vstack([data,new])#we keep the different points
    return data
```

### Here is how we keep points and all the points of each different  group.


The purpose of creating this function (choose_point) is that we want to keep point and keep all point of each different group

```python
def choose_point(data):
    #keep_point = to keep the point that different for each group
    #(for ex: there are 8 group that are different so we keep 8 point of x,y)
    keep_point = [tuple(data[5])] #we keep point in tuple for example [(52.0, 29.0), (23.0, 115.0)]
    keep_data = [[0,tuple(data[5])]]# we keep all point for each point that different in keep_point in tuple
    #(for ex: keep_data [[0, (52.0, 29.0), (53.0, 29.0),],[1, (174.0, 279.0), (174.0, 280.0),]])
    #we have 8 keep_point, so in keep_data we have 8 group of the point
    th = 40  # threshold
    current = 0
    for i in range(6,data.shape[0]-5):
        status = True
        for num in range(len(keep_point)):
            # cy,cx is for check that data point is in the threshold of some keep point or not
            cy = data[i] >= (keep_point[num][0]-th, keep_point[num][1]-th)
            cx = data[i] <= (keep_point[num][0]+th, keep_point[num][1]+th)
            if cx.all() and cy.all():  # if in that threshold 
                keep_data[num].append(tuple(data[i]))  # add to the list that match the keep point index
                status = False         
        if status:  # if there is a new set of keep point
            current += 1  # keep track how many set of keep point now
            keep_point.append(tuple(data[i]))
            keep_data.append([current,tuple(data[i])])
    return keep_point,keep_data
```

### This is where the function begin.


```python
def main(first_pic,second_pic):
    data = find_different_point(first_pic,second_pic)
    keep_point,keep_data = choose_point(data)
    plt.figure(figsize=(10,10))
    plt.imshow(second_blurred_half)
    #In this loop, the goal of it is that to find the mean of each keep_point in keep_data 
    for z in range(len(keep_data)):
        keep_value_y = 0.
        keep_value_x = 0.
        length = len(keep_data[z])
        for i in range(1,length):
            keepeiei = keep_data[z][i]
            keep_value_y+=keepeiei[0]
            keep_value_x+=keepeiei[1]
        keep_value_y = keep_value_y/(length-1)
        keep_value_x = keep_value_x/(length-1)
        
        #this is where i create the circle, then i plot it with the picture by using the mean point that i got
        circle2=plt.Circle((keep_value_y,keep_value_x),20, color = 'r', fill=False,linewidth = 5)
        fig = plt.gcf()
        fig.gca().add_artist(circle2)
    return keep_point
```
